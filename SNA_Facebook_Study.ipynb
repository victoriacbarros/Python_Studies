```python
G_fb = nx.read_edgelist('/Users/Victoria/Documents/Pos/Analytics_II/SNA_Python/facebook_combined.txt', create_using = nx.Graph(), nodetype = int)
```

```python
#summarizing datbase informations
print(nx.info(G_fb))
```

```python
#starting our graph analysis, viewing all the network existent
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb);
```

```python
#visualizing our netword according the size and node degree by the 'betweenness centrality' concept:

pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized =True, endpoints =True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size = [v*10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels= False,
               node_color=node_color,
               node_size=node_size)
plt.axis('off');
```

```python
#to know the labels of nodes with higher "betweenness centrality"
sorted(betCent, key=betCent.get, reverse=True)[:5]
```
